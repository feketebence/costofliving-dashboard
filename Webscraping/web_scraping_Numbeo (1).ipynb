# Categories

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
url = "https://www.numbeo.com/cost-of-living/in/Cluj-napoca"
page = urlopen(url)
html_bytes = page.read()
html = html_bytes.decode("utf-8")

soup = BeautifulSoup(html, "html.parser")
data = []
table = soup.find('table', attrs={'class':'data_wide_table new_bar_table'})

rows = table.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) # Get rid of empty values
    
data2 = [x for x in data if x != []]
categories = []
categories = [el[0] for el in data2]
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.numbeo.com/cost-of-living/rankings_current.jsp"
page = urlopen(url)
html_bytes = page.read()
html = html_bytes.decode("utf-8")
soup = BeautifulSoup(html, "html.parser")

cities_data = []
table = soup.find('table', attrs={'id':'t2'})
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    cities_data.append([ele for ele in cols if ele]) # Get rid of empty values

```

```python
cities_countries = [el[0]for el in cities_data]
```

```python
cities = []
for i in cities_countries:
    cities.append(i.split(", ")[0])
```

```python
def scraping_data():
    
    for city in cities:
        url = "https://www.numbeo.com/cost-of-living/in/"
        url = url + city + "?displayCurrency=EUR"
        print(url)
        page = urlopen(url)
        html_bytes = page.read()
        html = html_bytes.decode("utf-8")
        soup = BeautifulSoup(html, "html.parser")
        city_data = []
        table = soup.find('table', attrs={'class':'data_wide_table new_bar_table'})
        
        try:
            rows = table.find_all('tr')
        except:
            print("its no good: " + city)
            pass
        
        for row in rows:
            cols = row.find_all('td')
            cols = [ele.text.strip() for ele in cols]
            city_data.append([ele for ele in cols if ele])
            
        city_data2 = [x for x in city_data if x != []]  # Get rid of empty values
            
        prices_city = []
        prices_city = [el[1] for el in city_data2]
        
        df.loc[df['Cities'] == city, categories] = prices_city       
```
